```python
import json

dict_in_purchase = {}
i = 0
with open('purchase_log.txt', 'r', encoding='utf-8') as f:
    next (f)
    for line in f:
        purchase_1 = json.loads(line)
        dict_in_purchase[purchase_1['user_id']] = purchase_1['category']
        i += 1
        if i > 200:
            break    
```

```python
purchase_visit = str()
with open('visit_log.csv', 'r', encoding='utf-8') as f1:
    f2 = open ('funnel.csv', 'w')
    next (f1)
    for line in f1:
        line = line.strip('\n')
        line_1 = line.split(',')       
        line_2 =  dict_in_purchase.get(line_1[0])
        if line_2 != None:
            purchase_visit = line+','+line_2+'\n'
            print (purchase_visit)
            f2.write(purchase_visit)
        else:
            continue
f2.close()            
```
